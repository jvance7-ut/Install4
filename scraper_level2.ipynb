```python
from bs4 import BeautifulSoup #scraping
import requests #Open URL
import pandas as pd #dataframe
```

**Links**

```python
#making a GET request
r = requests.get('https://www.scrapethissite.com/pages/forms/?per_page=100')
r2 = requests.get('https://www.scrapethissite.com/pages/forms/?page_num=2&per_page=100')
r3 = requests.get('https://www.scrapethissite.com/pages/forms/?page_num=3&per_page=100')
r4 = requests.get('https://www.scrapethissite.com/pages/forms/?page_num=4&per_page=100')
r5 = requests.get('https://www.scrapethissite.com/pages/forms/?page_num=5&per_page=100')
r6 = requests.get('https://www.scrapethissite.com/pages/forms/?page_num=6&per_page=100')

#check status code for response received
#success code - 200
print(r, r2, r3, r4, r5, r6)
```

```python
#parsing the HTML
soup = BeautifulSoup(r.content, 'html.parser')
print(soup.prettify())

soup2 = BeautifulSoup(r2.content, 'html.parser')
print(soup2.prettify())

soup3 = BeautifulSoup(r3.content, 'html.parser')
print(soup3.prettify())

soup4 = BeautifulSoup(r4.content, 'html.parser')
print(soup4.prettify())

soup5 = BeautifulSoup(r5.content, 'html.parser')
print(soup5.prettify())

soup6 = BeautifulSoup(r6.content, 'html.parser')
print(soup6.prettify())
```

```python
#we want the table values
#the HTML value for our table is class_= "table"
#if you need help finding what values you want:
#1. go to the website
#2. right click and select 'Inspect' from the menu
#3. hover over the HTML until ONLY THE AREA YOU WANT is blue
#4. use the HTML tags from that line

table_names = soup.find('table', class_='table')

# Extract data from the tbody
if table_names:
    # Loop through rows
    for row in table_names.find_all('th'):
      #column names use 'th'
        print(row.text)
    #print(len(row.text))

else:
    print("raw_table not found.")


data_names = []
for row in table_names.find_all('th'):
    data_names.append(row.text.strip())


print("   ")
print(data_names)

print(len(data_names))
```

```python
# Find the table cell values with the specific class using tbody
tbody = soup.find('table', class_='table')
# Extract data from the tbody
if tbody:
    # Loop through rows
    for row in tbody.find_all('tr'):
        # Loop through cells
        for cell in row.find_all('td'):
            print(cell.text)
    #cell values use 'tr' and 'td'
else:
    print("Tbody not found.")


tbody2 = soup2.find('table', class_='table')
# Extract data from the tbody2
if tbody2:
    # Loop through rows
    for row in tbody2.find_all('tr'):
        # Loop through cells
        for cell in row.find_all('td'):
            print(cell.text)
    #cell values use 'tr' and 'td'
else:
    print("Tbody not found.")


tbody3 = soup3.find('table', class_='table')
# Extract data from the tbody3
if tbody3:
    # Loop through rows
    for row in tbody3.find_all('tr'):
        # Loop through cells
        for cell in row.find_all('td'):
            print(cell.text)
    #cell values use 'tr' and 'td'
else:
    print("Tbody not found.")


tbody4 = soup4.find('table', class_='table')
# Extract data from the tbody4
if tbody4:
    # Loop through rows
    for row in tbody4.find_all('tr'):
        # Loop through cells
        for cell in row.find_all('td'):
            print(cell.text)
    #cell values use 'tr' and 'td'
else:
    print("Tbody not found.")


tbody5 = soup5.find('table', class_='table')
# Extract data from the tbody5
if tbody5:
    # Loop through rows
    for row in tbody5.find_all('tr'):
        # Loop through cells
        for cell in row.find_all('td'):
            print(cell.text)
    #cell values use 'tr' and 'td'
else:
    print("Tbody not found.")


tbody6 = soup6.find('table', class_='table')
# Extract data from the tbody6
if tbody6:
    # Loop through rows
    for row in tbody6.find_all('tr'):
        # Loop through cells
        for cell in row.find_all('td'):
            print(cell.text)
    #cell values use 'tr' and 'td'
else:
    print("Tbody not found.")
```

**Turn data into dataframes**

```python
data = []
#empty data frame
for row in tbody.find_all('tr'):
    row_data = []
    for cell in row.find_all('td'):
        row_data.append(cell.text.strip())
    data.append(row_data)
#fill in empty dataframe row by row and column by column

#create a temporary dataframe for cleaning (optional)
df = pd.DataFrame(data)
df = df.drop(0)


data2 = []
#empty data frame
for row in tbody2.find_all('tr'):
    row_data = []
    for cell in row.find_all('td'):
        row_data.append(cell.text.strip())
    data2.append(row_data)
#fill in empty dataframe row by row and column by column

#create a temporary dataframe for cleaning (optional)
df2 = pd.DataFrame(data2)
df2 = df2.drop(0)


data3 = []
#empty data frame
for row in tbody3.find_all('tr'):
    row_data = []
    for cell in row.find_all('td'):
        row_data.append(cell.text.strip())
    data3.append(row_data)
#fill in empty dataframe row by row and column by column

#create a temporary dataframe for cleaning (optional)
df3 = pd.DataFrame(data3)
df3 = df3.drop(0)


data4 = []
#empty data frame
for row in tbody4.find_all('tr'):
    row_data = []
    for cell in row.find_all('td'):
        row_data.append(cell.text.strip())
    data4.append(row_data)
#fill in empty dataframe row by row and column by column

#create a temporary dataframe for cleaning (optional)
df4 = pd.DataFrame(data4)
df4 = df4.drop(0)


data5 = []
#empty data frame
for row in tbody5.find_all('tr'):
    row_data = []
    for cell in row.find_all('td'):
        row_data.append(cell.text.strip())
    data5.append(row_data)
#fill in empty dataframe row by row and column by column

#create a temporary dataframe for cleaning (optional)
df5 = pd.DataFrame(data5)
df5 = df5.drop(0)


data6 = []
#empty data frame
for row in tbody6.find_all('tr'):
    row_data = []
    for cell in row.find_all('td'):
        row_data.append(cell.text.strip())
    data6.append(row_data)
#fill in empty dataframe row by row and column by column

#create a temporary dataframe for cleaning (optional)
df6 = pd.DataFrame(data6)
df6 = df6.drop(0)


#print(df, df2, df3, df4, df5, df6)
```

```python
#assign the data_names as the column names
df.columns = data_names
df2.columns = data_names
df3.columns = data_names
df4.columns = data_names
df5.columns = data_names
df6.columns = data_names

#print(df, df2, df3, df4, df5, df6)
```

**Combine dataframes into one**

```python
combined_df = pd.concat([df, df2, df3, df4, df5, df6])

combined_df
```

```python
#Replace blank '' values with -1 to indicate rows that have no overtime losses
#There are other alternatives, but this allows clear distinction between no values and
#Positive values

print(combined_df['OT Losses'])
combined_df['OT Losses'] = combined_df['OT Losses'].replace('', '-1', regex=False)
print(combined_df['OT Losses'])
```

```python
##Change necessary columns to numeric.  Treating year as date/time value as we

combined_df['Year'] = pd.to_datetime(combined_df['Year'])
combined_df['Year'] = combined_df['Year'].dt.year
combined_df['Wins'] = pd.to_numeric(combined_df['Wins'])
combined_df['Losses'] = pd.to_numeric(combined_df['Losses'])
combined_df['OT Losses'] = pd.to_numeric(combined_df['OT Losses'])
combined_df['Win %'] = pd.to_numeric(combined_df['Win %'])
combined_df['Goals For (GF)'] = pd.to_numeric(combined_df['Goals For (GF)'])
combined_df['Goals Against (GA)'] = pd.to_numeric(combined_df['Goals Against (GA)'])
combined_df['+ / -'] = pd.to_numeric(combined_df['+ / -'])

##Summary statistics of numeric columns
print(combined_df.describe())
```
